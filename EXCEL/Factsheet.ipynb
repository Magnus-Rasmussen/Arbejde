# Forklaring af koden og hvordan API-kaldene kan ændres

Udvælgese af grupper

Hvis alle grupper ønskes, skriver der under filters:
"trb": ["1", "2", "3", "4", "5"],

Hvis kun overenskosmtansatte og tjenestemænd ønskes skrives der:
"trb": ["1", "2"],

For kun at vælge månedslønnede skrives der:
"afl": ["0", "1"],

Hvis dataet skal opdeles på Regioner skrives der i control:
"kom_reg"

Årene kan ændres som man vil, hvor der altid blot skal stå:
 {"y1": "Ønsket år", "m1": "Ønsket måned i tal"},

 så fx hvis marts 2025 ønskes skrives der:
 {"y1": "2025", "m1": "03"},

Hvis nogen variable ikke ønskes at medtages kan disse blot slettes under data selektionen. SÆrligt relevant i forhold til når det trækkes løndata. Hvor der fx blot kan udvælges:

"fuldtid","tl", "sy", "ov","ln","lb"

Hvis der ikke ønskes at have særydelser med og overarbejde ændrer man blot til:
"fuldtid","tl","ln","lb"
