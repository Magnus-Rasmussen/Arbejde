```python
from __future__ import annotations

from io import BytesIO
from pathlib import Path
import re
import zipfile
import xml.etree.ElementTree as ET

NS_MAIN = "http://schemas.openxmlformats.org/spreadsheetml/2006/main"
NS_REL = "http://schemas.openxmlformats.org/package/2006/relationships"
NS_APP = "http://schemas.openxmlformats.org/officeDocument/2006/extended-properties"
NS_VT = "http://schemas.openxmlformats.org/officeDocument/2006/docPropsVTypes"


def _p(tag: str) -> str:
    return f"{{{NS_MAIN}}}{tag}"


def _r(tag: str) -> str:
    return f"{{{NS_REL}}}{tag}"


def _a(tag: str) -> str:
    return f"{{{NS_APP}}}{tag}"


def _vt(tag: str) -> str:
    return f"{{{NS_VT}}}{tag}"


def _load_members(path: Path) -> dict[str, bytes]:
    with zipfile.ZipFile(path, "r") as z:
        return {name: z.read(name) for name in z.namelist()}


def _xml_bytes(tree: ET.ElementTree) -> bytes:
    return ET.tostring(tree.getroot(), encoding="utf-8", xml_declaration=True)


def _parse_xml(xml_bytes: bytes) -> ET.ElementTree:
    return ET.ElementTree(ET.fromstring(xml_bytes))


def _parse_shared_strings(xml_bytes: bytes) -> list[ET.Element]:
    root = ET.fromstring(xml_bytes)
    return list(root.findall(_p("si")))


def _count_shared_string_refs(sheet_xml: bytes) -> int:
    root = ET.fromstring(sheet_xml)
    count = 0
    for c in root.iter(_p("c")):
        if c.get("t") != "s":
            continue
        v = c.find(_p("v"))
        if v is None or v.text is None:
            continue
        count += 1
    return count


def _build_shared_strings_xml(si_list: list[ET.Element], total_ref_count: int) -> bytes:
    root = ET.Element(_p("sst"), {"count": str(total_ref_count), "uniqueCount": str(len(si_list))})
    for si in si_list:
        root.append(si)
    return ET.tostring(root, encoding="utf-8", xml_declaration=True)


def _remap_shared_string_indexes(sheet_xml: bytes, index_map: dict[int, int]) -> bytes:
    tree = _parse_xml(sheet_xml)
    root = tree.getroot()
    for c in root.iter(_p("c")):
        if c.get("t") != "s":
            continue
        v = c.find(_p("v"))
        if v is None or v.text is None:
            continue
        try:
            old_idx = int(v.text)
        except ValueError:
            continue
        new_idx = index_map.get(old_idx)
        if new_idx is not None:
            v.text = str(new_idx)
    return _xml_bytes(tree)


def _next_rid(existing_ids: set[str]) -> str:
    max_n = 0
    for rid in existing_ids:
        m = re.fullmatch(r"rId(\d+)", rid)
        if m:
            max_n = max(max_n, int(m.group(1)))
    return f"rId{max_n + 1}"


def _update_docprops_app(app_xml: bytes, sheet_names: list[str]) -> bytes:
    root = ET.fromstring(app_xml)

    heading_pairs = root.find(_a("HeadingPairs"))
    if heading_pairs is not None:
        vec = heading_pairs.find(_vt("vector"))
        if vec is not None:
            variants = vec.findall(_vt("variant"))
            if len(variants) >= 2:
                i4 = variants[1].find(_vt("i4"))
                if i4 is not None:
                    i4.text = str(len(sheet_names))

    titles = root.find(_a("TitlesOfParts"))
    if titles is not None:
        vec = titles.find(_vt("vector"))
        if vec is not None:
            vec.set("size", str(len(sheet_names)))
            for child in list(vec):
                vec.remove(child)
            for name in sheet_names:
                el = ET.Element(_vt("lpstr"))
                el.text = name
                vec.append(el)

    return ET.tostring(root, encoding="utf-8", xml_declaration=True)


def _update_workbook_xml_text(
    wb_xml: bytes,
    *,
    sheet_names: tuple[str, str, str],
    rid2: str,
    rid3: str,
) -> bytes:
    text = wb_xml.decode("utf-8")

    # Fix invalid attribute emitted by API export
    text = text.replace('state="show"', 'state="visible"')

    # Rename first sheet
    text2, _ = re.subn(r'(<sheet\b[^>]*\bname=")([^"]*)(")', rf'\1{sheet_names[0]}\3', text, count=1)
    text = text2

    sheet_ids = [int(x) for x in re.findall(r'\bsheetId="(\d+)"', text)]
    max_sheet_id = max(sheet_ids) if sheet_ids else 1

    # Insert 2 new <sheet .../> entries just before </sheets>
    m = re.search(r'(\n[\t ]*)</sheets>', text)
    indent = m.group(1) if m else "\n    "
    insertion = (
        f'{indent}<sheet name="{sheet_names[1]}" sheetId="{max_sheet_id + 1}" r:id="{rid2}"/>'
        f'{indent}<sheet name="{sheet_names[2]}" sheetId="{max_sheet_id + 2}" r:id="{rid3}"/>'
    )
    text = text.replace("</sheets>", insertion + "</sheets>", 1)

    return text.encode("utf-8")


def merge_three_xlsx_to_three_sheets(
    over_path: Path,
    stil_path: Path,
    klass_path: Path,
    out_path: Path,
    sheet_names: tuple[str, str, str] = ("overenskomst", "stilling", "klassificering"),
) -> None:
    base_members = _load_members(over_path)
    stil_members = _load_members(stil_path)
    klass_members = _load_members(klass_path)

    rels_tree = _parse_xml(base_members["xl/_rels/workbook.xml.rels"])
    rels_root = rels_tree.getroot()
    existing_rids = {rel.get("Id") for rel in rels_root.findall(_r("Relationship")) if rel.get("Id")}
    rid2 = _next_rid(existing_rids)
    existing_rids.add(rid2)
    rid3 = _next_rid(existing_rids)
    existing_rids.add(rid3)

    rel_type_ws = "http://schemas.openxmlformats.org/officeDocument/2006/relationships/worksheet"
    rels_root.append(ET.Element(_r("Relationship"), {"Id": rid2, "Type": rel_type_ws, "Target": "worksheets/sheet2.xml"}))
    rels_root.append(ET.Element(_r("Relationship"), {"Id": rid3, "Type": rel_type_ws, "Target": "worksheets/sheet3.xml"}))

    wb_xml_out = _update_workbook_xml_text(
        base_members["xl/workbook.xml"], sheet_names=sheet_names, rid2=rid2, rid3=rid3
    )

    base_sis = _parse_shared_strings(base_members["xl/sharedStrings.xml"]) if "xl/sharedStrings.xml" in base_members else []
    base_key_to_idx: dict[bytes, int] = {ET.tostring(si, encoding="utf-8"): i for i, si in enumerate(base_sis)}

    def merge_one_shared_strings(other_members: dict[str, bytes]) -> dict[int, int]:
        if "xl/sharedStrings.xml" not in other_members:
            return {}
        other_sis = _parse_shared_strings(other_members["xl/sharedStrings.xml"])
        idx_map: dict[int, int] = {}
        for old_i, si in enumerate(other_sis):
            key = ET.tostring(si, encoding="utf-8")
            existing = base_key_to_idx.get(key)
            if existing is not None:
                idx_map[old_i] = existing
                continue
            base_sis.append(si)
            new_i = len(base_sis) - 1
            base_key_to_idx[key] = new_i
            idx_map[old_i] = new_i
        return idx_map

    stil_map = merge_one_shared_strings(stil_members)
    klass_map = merge_one_shared_strings(klass_members)

    sheet2_xml = _remap_shared_string_indexes(stil_members["xl/worksheets/sheet1.xml"], stil_map)
    sheet3_xml = _remap_shared_string_indexes(klass_members["xl/worksheets/sheet1.xml"], klass_map)

    ct_ns = "http://schemas.openxmlformats.org/package/2006/content-types"
    ET.register_namespace("", ct_ns)
    ct_root = ET.fromstring(base_members["[Content_Types].xml"])

    def ct_tag(t: str) -> str:
        return f"{{{ct_ns}}}{t}"

    overrides = {ov.get("PartName") for ov in ct_root.findall(ct_tag("Override")) if ov.get("PartName")}
    ws_ct = "application/vnd.openxmlformats-officedocument.spreadsheetml.worksheet+xml"
    for part in ("/xl/worksheets/sheet2.xml", "/xl/worksheets/sheet3.xml"):
        if part not in overrides:
            ct_root.append(ET.Element(ct_tag("Override"), {"PartName": part, "ContentType": ws_ct}))

    app_xml_out = base_members.get("docProps/app.xml")
    if app_xml_out is not None:
        app_xml_out = _update_docprops_app(app_xml_out, list(sheet_names))

    base_sheet1_xml = base_members["xl/worksheets/sheet1.xml"]
    total_ref_count = _count_shared_string_refs(base_sheet1_xml) + _count_shared_string_refs(sheet2_xml) + _count_shared_string_refs(sheet3_xml)

    out_bytes_io = BytesIO()
    with zipfile.ZipFile(out_bytes_io, "w", compression=zipfile.ZIP_DEFLATED) as zout:
        for name, data in base_members.items():
            if name in ("xl/workbook.xml", "xl/_rels/workbook.xml.rels", "xl/sharedStrings.xml", "[Content_Types].xml", "docProps/app.xml"):
                continue
            zout.writestr(name, data)

        zout.writestr("xl/workbook.xml", wb_xml_out)
        zout.writestr("xl/_rels/workbook.xml.rels", _xml_bytes(rels_tree))
        if base_sis:
            zout.writestr("xl/sharedStrings.xml", _build_shared_strings_xml(base_sis, total_ref_count))
        zout.writestr("[Content_Types].xml", ET.tostring(ct_root, encoding="utf-8", xml_declaration=True))
        if app_xml_out is not None:
            zout.writestr("docProps/app.xml", app_xml_out)

        zout.writestr("xl/worksheets/sheet2.xml", sheet2_xml)
        zout.writestr("xl/worksheets/sheet3.xml", sheet3_xml)

    out_path.write_bytes(out_bytes_io.getvalue())


excel_dir = Path.cwd()
over = excel_dir / "arbedstid_overenskosmt.xlsx"
stil = excel_dir / "arbedstid_stilling.xlsx"
klass = excel_dir / "arbejdstid_klassificering.xlsx"
out_final = excel_dir / "arbjedstid.xlsx"
merge_three_xlsx_to_three_sheets(over, stil, klass, out_final)
print("Wrote:", out_final.resolve())
```
